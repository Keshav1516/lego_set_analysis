Lego Set Analysis

```python
import pandas as pd
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/lego_sets.csv")
theme=pd.read_csv("https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/parent_themes.csv")
```

```python
df.head()
theme.head(50)
```

```python
df.head()
```

```python
marged=df.merge(theme, left_on="parent_theme", right_on="name")
marged.head()
```

```python
marged.drop(columns="name_y", inplace=True)
marged.head()
```

```python
licensed= marged[marged["is_licensed"]]
licensed.head()

star_wars= licensed[licensed["parent_theme"]=="Star Wars"]
star_wars.head(50)
```

```python
star_wars.shape
```

```python
marged[marged["set_num"].isnull()].shape
```

```python
licensed=licensed.dropna(subset=["set_num"])
licensed.head()
```

```python
the_force=int(star_wars.shape[0]/licensed.shape[0]*100)
print(the_force)
```

```python
licensed_sorted= licensed.sort_values("year")
licensed_sorted.head()
```

```python
licensed_sorted["count"]=1
licensed_sorted.head()

summed_df=licensed_sorted.groupby(["year","parent_theme"]).sum().reset_index()
summed_df.head()
```

```python
max_df=summed_df.sort_values("count", ascending=False).drop_duplicates(["year"])
max_df.sort_values("year", inplace=True)
max_df
```

### Break down number of sets by year

```python
clean_df=marged[~marged["set_num"].isnull()]
clean_df["count"]=1
sets_per_years=clean_df.groupby(["year"]).sum().reset_index()[["year","count"]]

for index,row in sets_per_years.iterrows():
    print(row["year"],row["count"])
```
